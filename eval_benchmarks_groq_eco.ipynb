# ═════════════════════════════════════════════════════════════════════════════
# THE 3-TOPIC SAFE BENCHMARKING ENGINE
# ═════════════════════════════════════════════════════════════════════════════

```python
print("Initializing 3-Topic User Journey Simulation...")

dashboard_prompts = {
    "summary": "Based strictly on the provided parliamentary data, write a 3-paragraph executive summary.",
    "timeline": "Analyze the policy progression. Identify key years and format as a markdown timeline.",
    "ministry": "Identify which government ministries are engaged in this topic and summarize their responses.",
    "gaps": "Identify 'Policy Gaps'. What issues remain unresolved based on these records?"
}

dashboard_results = []
qa_results = []

# --- THE FIX: Slice the dictionary to only run the first 3 topics ---
topics_to_run = list(eval_suite.items())[:3]

# Update progress bar math to match the 3 topics
total_runs = len(topics_to_run) * len(models_to_test)
pbar = tqdm(total=total_runs, desc="Simulating User Journeys")

for topic_keyword, questions in topics_to_run:
    
    # ─── STEP 1: THE KEYWORD SEARCH ───
    broad_docs = ai_engine.retrieve(topic_keyword, top_k=4)
    dash_context = "\n---\n".join([d["text"] for d in broad_docs]) if broad_docs else "No specific records found."
    
    for model in models_to_test:
        
        # ─── STEP 2: DASHBOARD GENERATION (Parallel Load Test) ───
        def fetch_inference(prompt_type):
            sys_prompt = "You are an expert AI policy analyst. Ground inferences strictly in the text."
            user_prompt = f"{dashboard_prompts[prompt_type]}\n\nDATA:\n{dash_context}"
            try:
                resp = ai_engine.llm_client.chat.completions.create(
                    model=model,
                    messages=[{"role": "system", "content": sys_prompt}, {"role": "user", "content": user_prompt}],
                    max_tokens=1000, temperature=0.3
                )
                return resp.usage.total_tokens
            except Exception as e:
                print(f"Dashboard API Error ({model}): {e}")
                return 0

        dash_start = time.time()
        with concurrent.futures.ThreadPoolExecutor(max_workers=2) as executor:
            dash_tokens = list(executor.map(fetch_inference, dashboard_prompts.keys()))
        dash_latency = time.time() - dash_start
        
        dashboard_results.append({
            "Topic": topic_keyword,
            "Model": model,
            "Dashboard Wait Time (s)": round(dash_latency, 2),
            "Tokens Processed": sum(dash_tokens)
        })
        time.sleep(4) # Protect API Limits
        
        # ─── STEP 3: THE CHATBOT Q&A (Sequential Test) ───
        for q_idx, query in enumerate(questions):
            
            qa_docs = ai_engine.retrieve(query, top_k=3)
            qa_context = "\n---\n".join([d["text"] for d in qa_docs]) if qa_docs else "No specific records found."
            
            sys_prompt = f"""You are Sansad AI. You have access to these records: <context>{qa_context}</context>. 
            If the context has the answer, use it. If not, use general knowledge but explicitly state it is outside the database. Answer directly."""
            
            qa_start = time.time()
            try:
                resp = ai_engine.llm_client.chat.completions.create(
                    model=model,
                    messages=[{"role": "system", "content": sys_prompt}, {"role": "user", "content": query}],
                    max_tokens=800, temperature=0.1
                )
                qa_latency = time.time() - qa_start
                tps = resp.usage.total_tokens / qa_latency if qa_latency > 0 else 0
                answer = resp.choices[0].message.content
            except Exception as e:
                print(f"QA API Error ({model}): {e}")
                qa_latency, tps, answer = 0, 0, f"Error: {e}"
                
            qa_results.append({
                "Topic": topic_keyword,
                "Model": model,
                "Question": f"Q{q_idx+1}",
                "Latency (s)": round(qa_latency, 2),
                "Speed (TPS)": round(tps, 2),
                "Docs Found": len(qa_docs),
                "Answer Snippet": answer[:100] + "..."
            })
            time.sleep(4) # Protect API Limits
            
        pbar.update(1)

pbar.close()
print("Simulation Complete! You can now run Cell 4 to see the charts.")
```
